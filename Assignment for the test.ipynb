## Diffrent way to find the values

```python
# import packages
import PyPDF2
import re

# open the pdf file
object = PyPDF2.PdfFileReader("assignment.pdf")

# get number of pages
NumPages = object.getNumPages()

pageObj = pdfReader.getPage(6)
```

```python
# define keyterms
String = "Revenue"

# extract text and do the search
for i in range(0, NumPages):
    PageObj = object.getPage(i)
    print("this is page " + str(i)) 
    Text = PageObj.extractText() 
    # print(Text)
    ResSearch = re.search(String, Text)
    print(ResSearch)
```

```python
# define keyterms
String = "TOTAL EXPENSES"

# extract text and do the search
for i in range(0, NumPages):
    PageObj = object.getPage(i)
    print("this is page " + str(i)) 
    Text = PageObj.extractText() 
    # print(Text)
    ResSearch = re.search(String, Text)
    print(ResSearch)
```

## Extract all tables from PDF using Python

```python
import tabula

pdf_path = "https://www.bseindia.com/xml-data/corpfiling/AttachHis/0f58c131-c858-49a2-940c-6dc37ae935df.pdf"

dfs = tabula.read_pdf(pdf_path, pages='all')
```

```python
print(len(dfs))
```

```python
for i in range(len(dfs)):
    dfs[i].to_csv(f"table_{i}.csv")
```
